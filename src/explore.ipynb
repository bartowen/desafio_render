```python
df2= df.drop(columns=['Date','Activity Id','Route Name','Average Heart Rate (bpm)','Friend\'s Tagged','Notes','GPX File'])
df2
```

```python
df2.describe()
```

```python
def iqr(df,columna):
    estadisticas = df[columna].describe()
    rango_iq = estadisticas['75%'] - estadisticas['25%']
    lim_superior = estadisticas['75%'] + 1.5 * rango_iq
    lim_inferior = estadisticas['25%'] - 1.5 * rango_iq
    return lim_superior, lim_inferior

a,b = iqr(df2,'Calories Burned')

print(f'Limite superior: {a} y el limite inferior: {b}')
```
